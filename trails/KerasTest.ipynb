```python
group_0 = {
    "Acadia National Park", "Yosemite National Park", "Rocky Mountain", 
    "Great Smoky Mountains", "Mammoth Cave National Park", "Grand Teton National Park", 
    "Mount Rainer National Park", "Paradise"
}
group_1 = {
    "Grand Canyon National Park", "Petrified Forest", "Mesa Verde National Park", 
    "Arches National Park", "Zion National Park", "Badlands National Park", 
    "Death Valley National Park", "Big Bend National Park", "Bryce Canyon National Park", 
    "Canyonlands National Park", "Capitol Reef National Park", "Joshua Tree National Park"
}
group_2 = {
    "Glacier National Park", "North Cascades National Park", "Crater Lake National Park", 
    "Denali National Park and Preserve", "Kenai Fjords National Park", 
    "Wrangell-St Elias National Park and Preserve"
}
group_3 = {
    "Yellowstone National Park", "Shenandoah National Park", 
    "Sequoia and Kings Canyon National Park", "Olympic National Park", "Redwood National Park"
}
group_4 = {
    "Virgin Islands National Park", "Hawaii Volcanoes National Park", 
    "Everglades National Park", "Haleakala National Park", "Channel Islands", 
    "Dry Tortugas National Park"
}

park_categories = {park: 0 for park in group_0}
park_categories.update({park: 1 for park in group_1})
park_categories.update({park: 2 for park in group_2})
park_categories.update({park: 3 for park in group_3})
park_categories.update({park: 4 for park in group_4})

df['national_park'] = df['national_park'].map(park_categories)

df['national_park'] = pd.to_numeric(df['national_park'], downcast='integer', errors='coerce')

df = df.dropna(subset=['national_park'])

df['national_park'] = df['national_park'].astype(int)

print(df.head())
```

```python
import nltk
from nltk.corpus import stopwords
stops = stopwords.words('english')
```

```python
def make_dataset(data):
    data["comment_text"] = data["comment_text"].str.lower()
    data['comment_text'] = data['comment_text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stops)]))
    data = tf.data.Dataset.from_tensor_slices(
        (
            {
                "comment_text" : data[['comment_text']],
            },
            {
                "national_park" : data[['national_park']]
            }
        )
    )    
    data.batch(17)
    return data
```

```python
data = make_dataset(df)
```

```python
train_size = int(0.8*len(data))
val_size   = int(0.2*len(data))
```

```python
train = data.take(train_size)
val   = data.skip(train_size).take(val_size)
```

```python
from keras.layers import TextVectorization
import re
import string
size_vocabulary = 2000

def standardization(input_data):
    no_punctuation = tf.strings.regex_replace(input_data,
                                  '[%s]' % re.escape(string.punctuation),'')
    return no_punctuation 

review_vectorize_layer = TextVectorization(
    standardize=standardization,
    max_tokens=size_vocabulary, 
    output_mode='int',
    output_sequence_length=500) 

review_vectorize_layer.adapt(train.map(lambda x, y: x["comment_text"]))
```

```python
import keras
from keras import layers
```

```python
inputs = keras.Input(shape = (1,), name = "comment_text", dtype = 'string')
x = review_vectorize_layer(inputs)
x = layers.Embedding(size_vocabulary, output_dim=10, name="embedding")(x)
x = layers.GlobalAveragePooling1D()(x)
x = layers.Dense(64, activation="relu")(x)
x = layers.Dense(64, activation="relu")(x)
outputs = layers.Dense(5, name = "national_park")(x)
```

```python
model = keras.Model(inputs=inputs, outputs=outputs, name="park_model_1")
```

```python
model.summary()
```

```python
model.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=keras.optimizers.legacy.RMSprop(),
    metrics=["accuracy"]
)

#callback = keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)
history = model.fit(train, validation_data = val, epochs = 20, verbose = True)
```
